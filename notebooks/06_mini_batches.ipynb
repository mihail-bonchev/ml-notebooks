```python
import numpy as np

# Simulated dataset: 100 samples, 3 features each
np.random.seed(42)
X = np.random.randn(100, 3)
true_W = np.array([[1.5], [-2.0], [0.5]])
y = X @ true_W + np.random.randn(100, 1) * 0.3

print(X.shape)
print(y.shape)
print(true_W.shape)
```

```python
def get_batches(X, y, batch_size=16):
    n = len(X)
    indices = np.arange(n)
    np.random.shuffle(indices)
    for start in range(0, n, batch_size):
        end = start + batch_size
        batch_idx = indices[start:end]
        print(batch_idx)
        yield X[batch_idx], y[batch_idx]
```

```python
# Initialize weights and bias
W = np.random.randn(3, 1)
b = np.zeros((1,))

learning_rate = 0.05
epochs = 20
batch_size = 16

for epoch in range(epochs):
    total_loss = 0
    for Xb, yb in get_batches(X, y, batch_size):
        # Forward pass
        y_pred = Xb @ W + b
        
        # Loss (MSE)
        loss = np.mean((y_pred - yb) ** 2)
        total_loss += loss * len(Xb)
        
        # Gradients
        dloss_dypred = 2 * (y_pred - yb) / len(Xb)
        dW = Xb.T @ dloss_dypred
        db = np.sum(dloss_dypred)
        
        # Update
        W -= learning_rate * dW
        b -= learning_rate * db

    print(f"Epoch {epoch+1:02d} | Loss: {total_loss/len(X):.4f}")
```

```python
losses = []
for epoch in range(epochs):
    total_loss = 0
    for Xb, yb in get_batches(X, y, batch_size):
        y_pred = Xb @ W + b
        loss = np.mean((y_pred - yb) ** 2)
        total_loss += loss * len(Xb)
        dloss_dypred = 2 * (y_pred - yb) / len(Xb)
        dW = Xb.T @ dloss_dypred
        db = np.sum(dloss_dypred)
        W -= learning_rate * dW
        b -= learning_rate * db
    avg_loss = total_loss / len(X)
    losses.append(avg_loss)

import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Average Loss")
plt.title("Mini-batch training progress")
plt.show()
```
